```python
import json

def load_data(filename):
    with open(filename) as f:
        data = json.load(f)
    return data
```

```python
data =load_data("data1.json")
data
```

```python
def display_users(data):
    print("users and their connection:")
    for user in data['users']:
        print(f"ID.{user['id']}-{user['name']} is friends with {user['friends']} and have liked the pages {user['liked_pages']}")
    print("\nusers and their activities")
    for page in data['pages']:
        print(f"ID.{page['id']} is {page['name']}")
```

```python
display_users(data)
```
